```python
#code 2 
import tkinter as tk
import pickle

# Employee class
class Employee:
    def __init__(self, name, id_number, department, job_title, basic_salary):
        self.name = name
        self.id_number = id_number
        self.department = department
        self.job_title = job_title
        self.basic_salary = basic_salary

# Car class
class Car:
    def __init__(self, id_number, price, car_type):
        self.id_number = id_number
        self.price = price
        self.car_type = car_type

# Main window
class MainWindow(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self.title("Car Management System")

        # Create the main menu
        self.menu = tk.Menu(self)
        self.menu.add_command(label="Add Employee", command=self.open_add_employee)
        self.menu.add_command(label="Delete Employee", command=self.open_delete_employee)
        self.menu.add_command(label="Modify Employee", command=self.open_modify_employee)
        self.menu.add_command(label="Display Car Details", command=self.open_display_car_details)
        self.menu.add_command(label="Display Sales Details", command=self.open_display_sales_details)
        self.config(menu=self.menu)

    def open_add_employee(self):
        # Create a new window for adding an employee
        add_employee_window = tk.Toplevel(self)
        add_employee_window.title("Add Employee")

        # Create labels and entry fields for employee details
        label_name = tk.Label(add_employee_window, text="Name:")
        label_name.pack()
        entry_name = tk.Entry(add_employee_window)
        entry_name.pack()

        label_id = tk.Label(add_employee_window, text="ID Number:")
        label_id.pack()
        entry_id = tk.Entry(add_employee_window)
        entry_id.pack()

        label_department = tk.Label(add_employee_window, text="Department:")
        label_department.pack()
        entry_department = tk.Entry(add_employee_window)
        entry_department.pack()

        label_job_title = tk.Label(add_employee_window, text="Job Title:")
        label_job_title.pack()
        entry_job_title = tk.Entry(add_employee_window)
        entry_job_title.pack()

        label_basic_salary = tk.Label(add_employee_window, text="Basic Salary:")
        label_basic_salary.pack()
        entry_basic_salary = tk.Entry(add_employee_window)
        entry_basic_salary.pack()

        # Create a button to add the employee
        button_add_employee = tk.Button(add_employee_window, text="Add Employee",
                                        command=lambda: self.add_employee(entry_name.get(), entry_id.get(),
                                                                         entry_department.get(),
                                                                         entry_job_title.get(),
                                                                         entry_basic_salary.get()))
        button_add_employee.pack()

    def open_delete_employee(self):
        # Create a new window for deleting an employee
        delete_employee_window = tk.Toplevel(self)
        delete_employee_window.title("Delete Employee")

        # Create a label and entry field for the employee ID to delete
        label_id = tk.Label(delete_employee_window, text="ID Number:")
        label_id.pack()
        entry_id = tk.Entry(delete_employee_window)
        entry_id.pack()

        # Create a button to delete the employee
        button_delete_employee = tk.Button(delete_employee_window, text="Delete Employee",
                                           command=lambda: self.delete_employee(entry_id.get()))
        button_delete_employee.pack()

    def open_modify_employee(self):
        # Create a new window for modifying an employee
        modify_employee_window = tk.Toplevel(self)
        modify_employee_window.title("Modify Employee")
        # Create labels and entry fields for employee details
        label_id = tk.Label(modify_employee_window, text="ID Number:")
        label_id.pack()
        entry_id = tk.Entry(modify_employee_window)
        entry_id.pack()

        label_name = tk.Label(modify_employee_window, text="Name:")
        label_name.pack()
        entry_name = tk.Entry(modify_employee_window)
        entry_name.pack()

        label_department = tk.Label(modify_employee_window, text="Department:")
        label_department.pack()
        entry_department = tk.Entry(modify_employee_window)
        entry_department.pack()

        label_job_title = tk.Label(modify_employee_window, text="Job Title:")
        label_job_title.pack()
        entry_job_title = tk.Entry(modify_employee_window)
        entry_job_title.pack()

        label_basic_salary = tk.Label(modify_employee_window, text="Basic Salary:")
        label_basic_salary.pack()
        entry_basic_salary = tk.Entry(modify_employee_window)
        entry_basic_salary.pack()

        # Create a button to modify the employee
        button_modify_employee = tk.Button(modify_employee_window, text="Modify Employee",
                                           command=lambda: self.modify_employee(entry_id.get(), entry_name.get(),
                                                                                entry_department.get(),
                                                                                entry_job_title.get(),
                                                                                entry_basic_salary.get()))
        button_modify_employee.pack()

    def open_display_car_details(self):
        # Create a new window for displaying car details
        display_car_details_window = tk.Toplevel(self)
        display_car_details_window.title("Display Car Details")

        # Create a label and entry field for the car ID to display
        label_id = tk.Label(display_car_details_window, text="ID Number:")
        label_id.pack()
        entry_id = tk.Entry(display_car_details_window)
        entry_id.pack()

        # Create a button to display the car details
        button_display_car_details = tk.Button(display_car_details_window, text="Display Car Details",
                                               command=lambda: self.display_car_details(entry_id.get()))
        button_display_car_details.pack()

    def open_display_sales_details(self):
        # Create a new window for displaying sales details
        display_sales_details_window = tk.Toplevel(self)
        display_sales_details_window.title("Display Sales Details")

        # Create a label and entry field for the employee ID to display sales details
        label_id = tk.Label(display_sales_details_window, text="ID Number:")
        label_id.pack()
        entry_id = tk.Entry(display_sales_details_window)
        entry_id.pack()

        # Create a button to display the sales details
        button_display_sales_details = tk.Button(display_sales_details_window, text="Display Sales Details",
                                                 command=lambda: self.display_sales_details(entry_id.get()))
        button_display_sales_details.pack()

    def add_employee(self, name, id_number, department, job_title, basic_salary):
        # Create an Employee instance with the provided details
        employee = Employee(name, id_number, department, job_title, basic_salary)

        # Save the employee to a binary file using Pickle
        with open("employees.pkl", "ab") as file:
            pickle.dump(employee, file)

        # Show a success message
        tk.messagebox.showinfo("Success", "Employee added successfully.")

    def delete_employee(self, id_number):
        # Load the employees from the binary file
        employees = self.load_employees()

        # Find the employee with the provided ID number
        for employee in employees:
            if employee.id_number == id_number:
                # Remove the employee from the list
                employees.remove(employee)

                # Save the updated list back to the binary file
                self.save_employees(employees)

                # Show a success message
                tk.messagebox.showinfo
                tk.messagebox.showinfo("Success", "Employee deleted successfully.")
                return

        # Show an error message if the employee was not found
        tk.messagebox.showerror("Error", "Employee not found.")

    def modify_employee(self, id_number, name, department, job_title, basic_salary):
        # Load the employees from the binary file
        employees = self.load_employees()

        # Find the employee with the provided ID number
        for employee in employees:
            if employee.id_number == id_number:
                # Modify the employee details
                employee.name = name
                employee.department = department
                employee.job_title = job_title
                employee.basic_salary = basic_salary

                # Save the updated list back to the binary file
                self.save_employees(employees)

                # Show a success message
                tk.messagebox.showinfo("Success", "Employee modified successfully.")
                return

        # Show an error message if the employee was not found
        tk.messagebox.showerror("Error", "Employee not found.")

    def display_car_details(self, id_number):
        # Load the cars from the binary file
        cars = self.load_cars()

        # Find the car with the provided ID number
        for car in cars:
            if car.id_number == id_number:
                # Show the car details
                tk.messagebox.showinfo("Car Details", f"ID Number: {car.id_number}\nPrice: {car.price}\nType: {car.car_type}")
                return

        # Show an error message if the car was not found
        tk.messagebox.showerror("Error", "Car not found.")

    def display_sales_details(self, id_number):
        # Load the employees from the binary file
        employees = self.load_employees()

        # Find the employee with the provided ID number
        for employee in employees:
            if employee.id_number == id_number:
                # Show the sales details of the employee
                sales_details = ""
                for sale in employee.sales:
                    sales_details += f"Car Sold: {sale.car_name}\nSale Price: {sale.sale_price}\n\n"

                tk.messagebox.showinfo("Sales Details", sales_details)
                return

        # Show an error message if the employee was not found
        tk.messagebox.showerror("Error", "Employee not found.")

    def load_employees(self):
        # Load employees from the binary file
        try:
            with open("employees.pkl", "rb") as file:
                employees = []
                while True:
                    try:
                        employee = pickle.load(file)
                        employees.append(employee)
                    except EOFError:
                        break
                return employees
        except FileNotFoundError:
            return []

    def save_employees(self, employees):
        # Save employees to the binary file
        with open("employees.pkl", "wb") as file:
            for employee in employees:
                pickle.dump(employee, file)

    def load_cars(self):
        # Load cars from the binary file
        try:
            with open("cars.pkl", "rb") as file:
                cars = []
                while True:
                    try:
                        car = pickle.load(file)
                        cars.append(car)
                    except EOFError:
                        break
                return cars
        except FileNotFoundError:
            return []

    def save_cars(self, cars):
        # Save cars to the binary file
        with open("cars.pkl", "wb") as file:
            for car in cars:
                pickle.dump(car, file)

# Create the main window
if __name__ == "__main__":
    root = MainWindow()
    root.mainloop()
```
